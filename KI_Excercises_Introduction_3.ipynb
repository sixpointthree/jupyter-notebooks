```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

N = 10
s = 0.5 #strength of noise
x = np.linspace(0, 1, N)
noise = np.random.normal(0, s, N)
y = np.sin(2*np.pi*x) + noise
plt.scatter(x, y)

#for plotting
xx = np.linspace(0, 1, 100)
plt.plot(xx, np.sin(2*np.pi*xx))
```

```python
param_sets = {}
for K in range(1, N):
    param_sets[K] = np.polyfit(x, y, K)    
```

```python
for K, params in param_sets.items():
    print(f"K = {K}: Params: {params}")
    plt.plot(x, np.polyval(params, x))
```

```python
# Square error of this functions
for K, params in param_sets.items():
    y_actual = y
    y_predicted = np.polyval(params, x)
    MSE = np.square(np.subtract(y_predicted, y_actual)).mean()
    RMSE = np.sqrt(MSE)
    print(f"Root Mean Square Error for K = {K}: {RMSE}")
```
