```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from sympy import symbols, diff

x1 = np.arange(-6, 6, 1)
x2 = np.arange(-6, 6, 1)

X1, X2 = np.meshgrid(x1, x2)
nu = np.power(X1, 2) + 2 * np.power(X2, 2) - 2 * X1 - 2 * X1 * X2 +2
nu_flat = 2

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X1, X2, nu)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
