```python
import pandas as pd
```

```python
data = pd.read_csv("./vehicle_overland.csv", delimiter=";")
data # print
```

```python
import numpy as np

rpm, speed = data.RPM, data.speed
    
print(rpm)
print(speed)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

plt.plot(speed)
plt.show()
```

```python
# eliminate outlier
for i in range(speed.shape[0]):
    if speed[i] > 125:
        speed[i] = speed[i-1]
```

```python
x = np.linspace(0, len(speed) - 1, len(speed))
print(x)

speed_mean = np.mean(speed)
speed_min = np.min(speed)
speed_max = np.max(speed)
speed_std = np.std(speed)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero') # Setzt Aches auf y = 0
ax.spines['right'].set_color('none') # Entfernt Achse rechts
ax.spines['top'].set_color('none') # Entfernt Achse oben

ax.set_title(f"Mean: {speed_mean:.2f}, Min: {speed_min}, Max: {speed_max}, std deviation: {speed_std:.2f}")

plt.axhline(y=speed_mean, color="black")
plt.axhline(y=speed_mean-speed_std, color="gray")
plt.axhline(y=speed_mean+speed_std, color="gray")
plt.plot(speed)
```

```python
plt.hist(speed, density=1, bins=50)
plt.show()
```
