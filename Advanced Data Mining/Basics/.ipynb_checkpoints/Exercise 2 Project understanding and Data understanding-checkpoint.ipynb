```python
import pandas as pd
```

```python
df = pd.read_csv("wine.csv")
print(f"Features: {len(df.columns)}") 
print(df.columns)
print(f"Records: {len(df)}")
for i in range(0, len(df.columns)):
    print(df.iloc[i])
```

```python
# Drop column
df.drop('Proline', axis=1, inplace=True)
```

```python
# Drop duplicates
df.drop_duplicates(keep="first", inplace=True)
```

```python
# Drop NaNs 
df.dropna(inplace=True)
```

```python
# remove whitespaces: df.coluns.str.strip().str.replace(' ', '_')
# print types: df.dtypes
# find val: df.index[df.Ash.str.contains('SOMETHING')]
```

```python
# Drop invalid records due to wrong datatype
invalid_idx_list = []
for i in range(0, len(df)):
    try:
        float(df["Ash"].iloc[i])
    except ValueError as e:
        print(f"Idx[{i}]: {e}")
        #invalid_idx_list.append(i)
        #invalid_idx_list.append(i - 1)
        invalid_idx_list.append(i + 1)
for idx_to_drop in invalid_idx_list:
    df.drop(idx_to_drop, axis=0, inplace=True)
    print(f"Dropped line {idx_to_drop}")
ash = df["Ash"].astype('float')
```

```python
df
```

```python
# Visualize data
%matplotlib inline
import matplotlib.pyplot as plt
fig, axs = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(30,15))

axs[0].plot(df["Alcohol"], df.index, label="Alcohol")
axs[0].plot(df["Malic acid"], df.index, label="Malic acid")
axs[0].plot(df["Alcalinity of ash"], df.index, label="Alcalinity of ash")
axs[0].plot(df["Magnesium"], df.index, label="Magnesium")
axs[0].plot(df["Color intensity"], df.index, label="Color intensity")
axs[0].legend()
axs[0].grid(visible=True)

axs[1].plot(ash, df.index, label="Ash")
axs[1].plot(df["Total phenols"], df.index, label="Total phenols")
axs[1].plot(df["Flavanoids"], df.index, label="Flavanoids")
axs[1].plot(df["Nonflavanoid phenols"], df.index, label="Nonflavanoid phenols")
axs[1].plot(df["Proanthocyanins"], df.index, label="Proanthocyanins")
axs[1].plot(df["Hue"], df.index, label="Hue")
axs[1].plot(df["OD280/OD315 of diluted wines"], df.index, label="OD280/OD315 of diluted wines")
axs[1].legend()
axs[1].grid(visible=True)
```
